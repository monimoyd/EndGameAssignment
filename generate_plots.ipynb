```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eval_epoch_reward_df = pd.read_csv("results/eval_epoch_reward.csv")
```

```python
eval_epoch_reward_df
```

```python
eval_epoch_reward_df.columns=["episode", "total_reward"]
```

```python
fig, axs = plt.subplots(figsize=(10,10))
axs.plot(eval_epoch_reward_df["episode"], eval_epoch_reward_df["total_reward"], label="Evaluation After Training Episode vs Avearge Total Rewards")
axs.legend()
plt.xlabel("Episode")
plt.ylabel("Average Total Rewards")
plt.show()
```

```python
eval_on_road_stats_df = pd.read_csv("results/eval_on_road_stats.csv")
```

```python
eval_on_road_stats_df.columns=['train_episode','eval_episode', 'on_road_count', 'off_road_count', 'hit_boundary_count', 'hit_goal_count']
```

```python
eval_on_road_stats_df.head()
```

```python
eval_on_road_stats_df["on_road_percentage"]=eval_on_road_stats_df["on_road_count"]/(eval_on_road_stats_df["on_road_count"] + eval_on_road_stats_df["off_road_count"])*100
```

```python
eval_on_road_stats_df.head()
```

```python
eval_on_road_stats_filtered_df = eval_on_road_stats_df[["train_episode", "on_road_percentage", "hit_boundary_count", "hit_goal_count"]]
```

```python
eval_on_road_stats_filtered_mean_df = eval_on_road_stats_filtered_df.groupby(['train_episode']).mean()
```

```python
eval_on_road_stats_filtered_mean_df.head()
```

```python
fig, axs = plt.subplots(figsize=(10,10))
axs.plot(eval_on_road_stats_filtered_mean_df.index,eval_on_road_stats_filtered_mean_df["on_road_percentage"], label="Evaluation After Training Episode vs On Road Percentage")
axs.legend()
plt.xlabel("Episode")
plt.ylabel("On Road Percentage")
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(10,10))
axs.plot(eval_on_road_stats_filtered_mean_df.index,eval_on_road_stats_filtered_mean_df["hit_boundary_count"], label="Evaluation After Training Episode vs Average Hit Boundary")
axs.legend()
plt.xlabel("Episode")
plt.ylabel("Average Hit Boundary")
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(10,10))
axs.plot(eval_on_road_stats_filtered_mean_df.index,eval_on_road_stats_filtered_mean_df["hit_goal_count"], label="Evaluation After Training Episode vs Average Hit Goal")
axs.legend()
plt.xlabel("Episode")
plt.ylabel("Average Hit Goal")
plt.show()
```
